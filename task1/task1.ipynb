```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'osn.csv'
data = pd.read_csv(file_path)
```

```python
sns.set(style="whitegrid")

# Bar chart for gender distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Gender", palette="pastel")
plt.title("Distribution of Genders", fontsize=16)
plt.xlabel("Gender", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="Kelas", kde=True)
plt.title("Distribution of Kelas (Class Levels)", fontsize=16)
plt.xlabel("Kelas", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.show()
```
